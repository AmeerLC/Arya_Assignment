## **Exploratory Data Analysis of Training Dataset**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"/content/training_set.csv")
```

```python
df.drop("Unnamed: 0", axis =1 , inplace = True)
df.shape
```

The dataset consist of 58 features and 3910 observations in the training dataset

```python
df.head(10)
```

```python
sparsity = 1 - (np.count_nonzero(df)/df.size)
sparsity
```

```python
df.describe()
```

```python
# To plot pie chart of output distribution
f, ax = plt.subplots(1, figsize = (18, 8))
df["Y"].value_counts().plot.pie(ax = ax, autopct = '%1.1f%%')
ax.set_title("Output distribution")   
sns.countplot(data = df, ax = ax[1])
plt.show()
```

The distribution of the output variable is 60 to 40

```python
columns = df.columns

# plotting histogram
for i in columns[:57]:
    f,ax=plt.subplots(figsize=(18,8))
    df[i].plot.hist(ax=ax)
    ax.set_title(i)
    plt.show()
```

The data variable are non normal and highly skewed.

```python
#Correlogram
sns.heatmap(df.corr(),annot=True,cmap='RdYlGn') 
fig=plt.gcf()
fig.set_size_inches(60,20)
plt.show()
```
